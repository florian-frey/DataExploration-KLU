```python
import pandas as pd
import json
```

```python
response = "api_responses\\20220615-144242_grand theft auto.json"
data = json.load(open(response))
```

```python
tweets = pd.DataFrame(data["data"])
tweets = tweets[["author_id", "id", "source", "created_at", "text"]]
# drop tweets by bots:
tweets = tweets.loc[(tweets["source"] != "Microsoft Power Platform")
                    & (tweets["source"] != "WordPress.com")
                    & (tweets["source"].str.contains("bot") == False)
                    & (tweets["source"].str.contains("Bot") == False)]
tweets.insert(4, "binary_label", "")
tweets
```
